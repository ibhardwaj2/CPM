**Scaling**

```python
# Import necessary libraries
from sklearn.preprocessing import StandardScaler
```

```python
# Define the columns to scale
columns_to_scale = ['Mat', 'Inns', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0']

```

```python
# Initialize the StandardScaler
scaler = StandardScaler()
```

```python
# Fit the scaler and transform the data for batting_data
batting_data[columns_to_scale] = scaler.fit_transform(batting_data[columns_to_scale])
```

```python
# Define the columns to scale for bowling_data
columns_to_scale_bowling = ['Mat', 'Inns', 'Balls', 'Runs', 'Wkts', 'Ave', 'Econ', 'SR', '4', '5', 'Best_Bowling_Wickets', 'Best_Bowling_Runs']

# Fit the scaler and transform the data for bowling_data
bowling_data[columns_to_scale_bowling] = scaler.fit_transform(bowling_data[columns_to_scale_bowling])
```

```python
# Fit the scaler and transform the data for fielding_data
columns_to_scale = ['Mat', 'Inns', 'Dis', 'Ct', 'St', 'Ct Wk', 'Ct Fi', 'D/I']
fielding_data[columns_to_scale] = scaler.fit_transform(fielding_data[columns_to_scale])
```

```python
# creating labels based on Batting_Average and Batting_Strike_Rate

def label_batting_data(row):
    if row['Batting_Average'] > 40 and row['Batting_Strike_Rate'] > 80:
        return 'Buy'
    elif row['Batting_Average'] < 30 and row['Batting_Strike_Rate'] < 70:
        return 'Sell'
    else:
        return 'Hold'

# Apply the labeling function to create labels
batting_data['Bet_Labels'] = batting_data.apply(label_batting_data, axis=1)

# Display the updated batting_data with labels
batting_data.head()
```

```python
from sklearn.model_selection import train_test_split

# Define the features (X) and labels (y)
features = batting_data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', 'Career_Duration', 'Batting_Average', 'Batting_Strike_Rate', 'Batting_Conversion_Rate']]
labels = batting_data['Bet_Labels']

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Display the shapes of the resulting sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```
